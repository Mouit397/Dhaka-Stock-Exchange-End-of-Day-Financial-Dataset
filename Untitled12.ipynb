```python
# =============================================================================
# PART 1: DATA LOADING
# =============================================================================

def load_single_stock(ticker: str, data_dir: str) -> pd.DataFrame | None:
    """
    Load ONE company's CSV from the adjusted-data folder.

    The dataset stores one CSV per company.
    Filename = ticker  (BRACBANK.csv → ticker BRACBANK).
    Columns  = Date, Open, High, Low, Close, Volume  — already adjusted.
    There is NO ticker column inside the file.

    Returns a long-format DataFrame with TICKER and DAILY_RETURN added,
    or None if the file does not exist.
    """
    path = os.path.join(data_dir, f"{ticker}.csv")
    if not os.path.exists(path):
        return None

    df = pd.read_csv(path, parse_dates=['Date'])
    df.rename(columns={
        'Date'  : 'DATE',
        'Open'  : 'OPEN',
        'High'  : 'HIGH',
        'Low'   : 'LOW',
        'Close' : 'CLOSE',
        'Volume': 'VOLUME',
    }, inplace=True)

    df['TICKER']       = ticker
    df['SECTOR']       = get_sector(ticker)
    df['DAILY_RETURN'] = df['CLOSE'].pct_change()
    df['LOG_RETURN']   = np.log(df['CLOSE'] / df['CLOSE'].shift(1))

    # Remove duplicate dates (rare but possible in source data)
    df = df.sort_values('DATE').drop_duplicates('DATE')
    return df


def load_dse_data(sector_map: dict = SECTOR_MAP,
                  data_dir: str   = DATA_DIR,
                  start_date: str = START_DATE) -> pd.DataFrame:
    """
    Load ALL companies listed in sector_map into one long-format DataFrame.

    Returns
    -------
    df : pd.DataFrame
        Columns: DATE, TICKER, SECTOR, OPEN, HIGH, LOW, CLOSE, VOLUME,
                 DAILY_RETURN, LOG_RETURN
        One row per (company, trading day).
    """
    all_tickers = [t for tickers in sector_map.values() for t in tickers]
    frames = []
    skipped = []

    print(f"Loading {len(all_tickers)} tickers from:\n  {data_dir}\n")

    for ticker in sorted(set(all_tickers)):
        df = load_single_stock(ticker, data_dir)
        if df is not None and len(df) > 200:
            df = df[df['DATE'] >= start_date]
            frames.append(df)
        else:
            skipped.append(ticker)

    if not frames:
        raise FileNotFoundError(
            f"No CSV files loaded from {data_dir}\n"
            "Check that DATA_DIR points to the 'Company Separated Adjusted Data' folder."
        )

    long_df = (pd.concat(frames, ignore_index=True)
                 .sort_values(['TICKER', 'DATE'])
                 .reset_index(drop=True))

    print(f"✓ Loaded  : {long_df['TICKER'].nunique()} stocks  ×  "
          f"{long_df['DATE'].nunique()} trading days")
    print(f"✓ Rows    : {len(long_df):,}")
    print(f"✓ Dates   : {long_df['DATE'].min().date()}  →  {long_df['DATE'].max().date()}")
    if skipped:
        print(f"⚠ Skipped : {len(skipped)} tickers (file not found): {skipped[:10]}")

    return long_df


# Convenience: load just the DSEX index for benchmark comparisons
def load_dsex(data_dir: str = DATA_DIR) -> pd.DataFrame:
    """Load the DSEX broad market index."""
    df = load_single_stock('00DSEX', data_dir)
    if df is None:
        print("⚠ 00DSEX.csv not found — skipping benchmark")
        return None
    return df[df['DATE'] >= START_DATE]
```

```python
# =============================================================================
# PART 2: HELPER FUNCTIONS
# =============================================================================

def max_drawdown(price_series: pd.Series) -> float:
    """
    Worst peak-to-trough decline.
    Returns a negative fraction, e.g. -0.82 means lost 82% from peak.
    """
    p = price_series.dropna()
    if len(p) < 2:
        return np.nan
    return float(((p - p.cummax()) / p.cummax()).min())


def sharpe_ratio(return_series: pd.Series, rf_daily: float = 0.0) -> float:
    """
    Annualised Sharpe = (mean excess return / std) × √252
    > 0.5 = good   |   > 1.0 = excellent   |   < 0 = avoid
    """
    r = return_series.dropna()
    if r.std() == 0 or len(r) < 50:
        return np.nan
    return float((r.mean() - rf_daily) / r.std() * np.sqrt(252))


def build_sector_index(df: pd.DataFrame) -> pd.DataFrame:
    """
    Build an equal-weighted sector price index.

    Each stock is normalised to 1.0 at its first date so that
    different price levels (BRACBANK ≈ 67 vs ABBANK ≈ 4) do not bias
    the average.  Returns a DataFrame:  index=DATE, columns=SECTOR.
    """
    pivot = df.pivot_table(index='DATE', columns='TICKER', values='CLOSE')
    normed = pivot.div(pivot.bfill().iloc[0])        # normalise to 1.0

    sector_idx = {}
    for sector, tickers in SECTOR_MAP.items():
        cols = [t for t in tickers if t in normed.columns]
        if cols:
            sector_idx[sector] = normed[cols].mean(axis=1)

    return pd.DataFrame(sector_idx).sort_index()
```

```python
# =============================================================================
# PART 3: TASK 1 — EXPLORATORY ANALYSIS & INVESTMENT INSIGHTS
# =============================================================================

def task1_exploratory_analysis(df: pd.DataFrame) -> dict:
    """
    Task 1: Open-ended findings with investment-relevant insights.

    Covers
    ------
    1.1  Market overview
    1.2  Return distribution & Sharpe ranking
    1.3  Sector volatility
    1.4  Multi-horizon momentum (1M / 3M / 6M / 1Y)
    1.5  Volume anomaly detection
    1.6  Maximum drawdown (resilience)
    1.7  Cross-stock correlation & pair-trading candidates
    1.8  Beta vs equal-weighted market
    """
    print("\n" + "="*70)
    print("TASK 1: EXPLORATORY ANALYSIS & INVESTMENT INSIGHTS")
    print("="*70)

    # ── 1.1 Market Overview ───────────────────────────────────────────────
    print("\n--- 1.1 MARKET OVERVIEW ---")
    print(f"  Trading days  : {df['DATE'].nunique()}")
    print(f"  Companies     : {df['TICKER'].nunique()}")
    print(f"  Sectors       : {df['SECTOR'].nunique()}")
    sector_counts = (df.drop_duplicates('TICKER')['SECTOR']
                       .value_counts()
                       .rename('Companies'))
    print(f"\n  Sector distribution:\n{sector_counts.to_string()}")

    # ── 1.2 Sharpe Ratio Ranking ──────────────────────────────────────────
    print("\n--- 1.2 RETURN DISTRIBUTION & SHARPE RANKING ---")
    stats = (df.groupby('TICKER')['DAILY_RETURN']
               .agg(mean='mean', std='std', count='count')
               .reset_index())
    stats['ANN_RETURN']  = (1 + stats['mean'])**252 - 1
    stats['ANN_VOL']     = stats['std'] * np.sqrt(252)
    stats['SHARPE']      = stats['ANN_RETURN'] / stats['ANN_VOL']
    stats['SECTOR']      = stats['TICKER'].apply(get_sector)
    stats = stats.sort_values('SHARPE', ascending=False)

    print("\n  Top 10 by Sharpe ratio (best risk-adjusted return):")
    print(stats.head(10)[['TICKER','SECTOR','SHARPE','ANN_RETURN','ANN_VOL']]
                .round(3).to_string(index=False))

    print("\n  Bottom 10 by Sharpe ratio (avoid — wealth-destroying):")
    print(stats.tail(10)[['TICKER','SECTOR','SHARPE','ANN_RETURN','ANN_VOL']]
                .round(3).to_string(index=False))

    # ── 1.3 Sector Volatility ─────────────────────────────────────────────
    print("\n--- 1.3 SECTOR VOLATILITY ---")
    vol_by_sector = (df.groupby('SECTOR')['DAILY_RETURN']
                       .std()
                       .mul(np.sqrt(252) * 100)
                       .sort_values(ascending=False)
                       .rename('Ann_Vol_%'))
    print(f"\n  Annualised volatility by sector (%):\n{vol_by_sector.round(1).to_string()}")

    # ── 1.4 Multi-Horizon Momentum ────────────────────────────────────────
    print("\n--- 1.4 MOMENTUM INSIGHTS ---")
    latest_date = df['DATE'].max()
    # Map horizon name to approximate trading-day count
    horizons = {'1M': 21, '3M': 63, '6M': 126, '1Y': 252}

    momentum_results = {}
    pivot_close = df.pivot_table(index='DATE', columns='TICKER', values='CLOSE')

    for label, days in horizons.items():
        # Use actual last N trading days (not calendar days)
        available_dates = pivot_close.index.sort_values()
        if len(available_dates) < days:
            continue
        start_date = available_dates[-days]
        window = pivot_close.loc[start_date:]
        cum_ret = (window.iloc[-1] / window.iloc[0] - 1) * 100
        cum_ret = cum_ret.dropna().sort_values(ascending=False)
        momentum_results[label] = cum_ret
        print(f"\n  Top 5 performers ({label}):")
        print(cum_ret.head(5).round(1).to_string())

    # ── 1.5 Volume Anomalies ──────────────────────────────────────────────
    print("\n--- 1.5 VOLUME ANOMALIES (Money-flow signals) ---")
    # Compare last 20-day avg volume to full-history avg volume
    recent_vol = (df[df['DATE'] >= df['DATE'].max() - pd.Timedelta(days=30)]
                    .groupby('TICKER')['VOLUME'].mean()
                    .rename('RECENT_VOL'))
    hist_vol   = df.groupby('TICKER')['VOLUME'].mean().rename('HIST_VOL')
    vol_signal = pd.concat([recent_vol, hist_vol], axis=1).dropna()
    vol_signal['SURGE'] = vol_signal['RECENT_VOL'] / vol_signal['HIST_VOL']
    vol_signal = vol_signal.sort_values('SURGE', ascending=False)

    print("\n  Stocks with recent volume surge (recent / historical avg):")
    print(vol_signal.head(15)[['SURGE']].round(2).to_string())
    print("\n  Interpretation: SURGE > 1.5 = unusual activity, potential signal")

    # ── 1.6 Max Drawdown ──────────────────────────────────────────────────
    print("\n--- 1.6 MAX DRAWDOWN (Resilience) ---")
    dd = (df.groupby('TICKER')['CLOSE']
            .apply(max_drawdown)
            .sort_values()
            .mul(100)
            .rename('MaxDD_%'))

    print("\n  Most resilient stocks (smallest max drawdown):")
    print(dd.head(10).round(1).to_string())
    print("\n  Highest-risk stocks (largest max drawdown):")
    print(dd.tail(10).round(1).to_string())

    # ── 1.7 Correlation & Pair-Trading ────────────────────────────────────
    print("\n--- 1.7 CROSS-STOCK CORRELATION ---")
    pivot_ret = df.pivot_table(index='DATE', columns='TICKER', values='DAILY_RETURN')
    corr_matrix = pivot_ret.corr()

    pairs = [
        (c1, c2, corr_matrix.loc[c1, c2])
        for i, c1 in enumerate(corr_matrix.columns)
        for c2 in corr_matrix.columns[i+1:]
        if abs(corr_matrix.loc[c1, c2]) < 1.0   # exclude self
    ]
    pairs.sort(key=lambda x: abs(x[2]), reverse=True)

    print("\n  Top 10 most correlated pairs (diversification warning):")
    for t1, t2, c in pairs[:10]:
        print(f"    {t1:15s} ↔ {t2:15s}: r = {c:+.3f}")

    print("\n  Top 10 least correlated pairs (best for diversification):")
    for t1, t2, c in sorted(pairs, key=lambda x: abs(x[2]))[:10]:
        print(f"    {t1:15s} ↔ {t2:15s}: r = {c:+.3f}")

    # ── 1.8 Beta ──────────────────────────────────────────────────────────
    print("\n--- 1.8 BETA vs EQUAL-WEIGHTED MARKET ---")
    market_ret = pivot_ret.mean(axis=1)     # equal-weighted market proxy

    betas = {}
    for ticker in pivot_ret.columns:
        s = pivot_ret[ticker].dropna()
        common = s.index.intersection(market_ret.index)
        if len(common) > 50:
            cov = np.cov(s[common], market_ret[common])[0, 1]
            var = market_ret[common].var()
            betas[ticker] = cov / var if var > 0 else np.nan

    beta_df = (pd.DataFrame.from_dict(betas, orient='index', columns=['BETA'])
                 .sort_values('BETA', ascending=False))

    print("\n  High-beta stocks (aggressive — amplify market moves):")
    print(beta_df.head(10).round(3).to_string())
    print("\n  Low-beta stocks (defensive — buffer against market drops):")
    print(beta_df.tail(10).round(3).to_string())

    return {
        'sharpe_stats'      : stats,
        'vol_by_sector'     : vol_by_sector,
        'momentum'          : momentum_results,
        'vol_signal'        : vol_signal,
        'max_drawdown'      : dd,
        'correlation_pairs' : pairs,
        'betas'             : beta_df,
    }
```

```python
# =============================================================================
# PART 4: TASK 2 — SECTOR ROTATION STRATEGY
# =============================================================================

def task2_sector_rotation(df: pd.DataFrame,
                           lookback_weeks: int = 4,
                           hold_weeks: int     = 4) -> dict:
    """
    Task 2: Sector Rotation Strategy — how money moves between sectors.

    Three signals
    -------------
    1. Relative momentum  : sector return vs market over lookback_weeks
    2. Volume surge ratio : recent volume / 6-month rolling average
    3. Composite score    : average of normalised signals → BUY / WATCH / AVOID

    Sections
    --------
    2.1  Build sector indices (equal-weighted)
    2.2  Historical sector momentum rankings
    2.3  Momentum persistence test (does past winner keep winning?)
    2.4  Volume surge / money flow
    2.5  Composite rotation signal
    2.6  Rotation pattern detection (leadership transitions)
    2.7  Prediction for next period
    """
    print("\n" + "="*70)
    print("TASK 2: SECTOR ROTATION STRATEGY")
    print("="*70)
    print(f"\n  Parameters: lookback = {lookback_weeks}w, hold = {hold_weeks}w")

    # ── 2.1 Sector Indices ────────────────────────────────────────────────
    sec_idx = build_sector_index(df)           # DATE × SECTOR price index
    sec_ret = sec_idx.pct_change()             # monthly sector returns
    market  = sec_ret.mean(axis=1)             # equal-weighted market

    # Monthly resampling for cleaner signals
    sec_monthly = (sec_idx
                   .resample('ME').last()
                   .pct_change()
                   .dropna(how='all'))

    # ── 2.2 Historical Rankings ───────────────────────────────────────────
    print("\n--- 2.2 HISTORICAL SECTOR MOMENTUM RANKINGS ---")
    lookback_days = lookback_weeks * 5   # trading days

    rebalance_dates = pd.date_range(
        start=sec_ret.index[lookback_days],
        end=sec_ret.index[-1],
        freq=f'{hold_weeks}W'
    )

    def relative_momentum(end_date, n_days):
        """Sector cumulative return minus market return over last n_days."""
        window = sec_ret.loc[:end_date].tail(n_days)
        sec_cum  = (1 + window.fillna(0)).prod() - 1
        mkt_cum  = (1 + market.loc[:end_date].tail(n_days).fillna(0)).prod() - 1
        return (sec_cum - mkt_cum).sort_values(ascending=False)

    leadership_history = []
    for reb_date in rebalance_dates:
        mom = relative_momentum(reb_date, lookback_days)
        leadership_history.append(mom.index[0])  # sector with best momentum

    # Show last 5 rebalancing signals
    print(f"\n  Showing last 5 rebalancing signals:")
    for reb_date in rebalance_dates[-5:]:
        mom = relative_momentum(reb_date, lookback_days)
        print(f"\n  {reb_date.strftime('%Y-%m-%d')}")
        print("    BUY  → " + ", ".join(f"{s}({v:+.1%})" for s,v in mom.head(3).items()))
        print("    AVOID→ " + ", ".join(f"{s}({v:+.1%})" for s,v in mom.tail(3).items()))

    # ── 2.3 Momentum Persistence ──────────────────────────────────────────
    print("\n--- 2.3 MOMENTUM PERSISTENCE TEST ---")
    print("  Does last period's top sector stay on top next period?")

    correlations = []
    for i in range(len(rebalance_dates) - 1):
        d_now  = rebalance_dates[i]
        d_next = rebalance_dates[i + 1]
        past_mom   = relative_momentum(d_now,  lookback_days)
        future_ret = (sec_ret.loc[d_now:d_next].fillna(0)
                              .apply(lambda c: (1+c).prod()-1))
        common = past_mom.index.intersection(future_ret.index)
        if len(common) >= 3:
            past_rank = past_mom[common].rank(ascending=False)
            corr = np.corrcoef(past_rank, future_ret[common])[0, 1]
            correlations.append(corr)

    persistence = np.mean(correlations) if correlations else 0
    print(f"\n  Average rank correlation (past→future): {persistence:+.3f}")
    if persistence > 0.1:
        print("  => MOMENTUM PERSISTENCE confirmed — strategy has predictive power!")
    elif persistence > 0:
        print("  => Weak momentum — useful with additional filters")
    else:
        print("  => Mean reversion — consider inverse (contrarian) strategy")

    # ── 2.4 Volume Surge / Money Flow ─────────────────────────────────────
    print("\n--- 2.4 VOLUME SURGE & MONEY FLOW ---")
    print("  Volume surge = recent monthly volume / 6-month rolling average")
    print("  > 1.5 = money flowing IN  |  < 0.7 = money flowing OUT")

    vol_monthly = (
    df.groupby(['TICKER', pd.Grouper(key='DATE', freq='ME')])['VOLUME']
    .sum()
    .reset_index()
    .rename(columns={'VOLUME': 'VOL_MONTHLY'})
)

    # Average across tickers in each sector
    vol_monthly['SECTOR'] = vol_monthly['TICKER'].apply(get_sector)
    sec_vol = (vol_monthly.groupby(['DATE','SECTOR'])['VOL_MONTHLY']
                           .mean()
                           .reset_index())

    sec_vol_pivot = sec_vol.pivot(index='DATE', columns='SECTOR', values='VOL_MONTHLY')
    vol_ma6       = sec_vol_pivot.rolling(6, min_periods=3).mean()
    vol_surge     = sec_vol_pivot / vol_ma6

    print("\n  Sector volume surge ratio (last 6 months):")
    print(vol_surge.tail(6).round(2).to_string())

    # Total money flow (volume-weighted return) by sector
    df['VW_RETURN'] = df['DAILY_RETURN'] * df['VOLUME']
    money_flow = (df.groupby('SECTOR')['VW_RETURN']
                    .sum()
                    .sort_values(ascending=False)
                    .rename('Vol_Wtd_Return'))
    print("\n  Cumulative volume-weighted return by sector (proxy for smart money):")
    print(money_flow.to_string())

    # ── 2.5 Composite Rotation Score ──────────────────────────────────────
    print("\n--- 2.5 COMPOSITE ROTATION SIGNAL ---")

    latest_mom    = relative_momentum(sec_ret.index[-1], lookback_days)
    latest_surge  = vol_surge.iloc[-1] if not vol_surge.empty else pd.Series()

    def minmax(s):
        s = s.dropna()
        r, mx = s.min(), s.max()
        return (s - r) / (mx - r) if mx > r else s * 0 + 0.5

    common_sectors = list(set(latest_mom.index) & set(latest_surge.index))
    if common_sectors:
        score = pd.DataFrame({
            'Momentum' : minmax(latest_mom[common_sectors]),
            'Volume'   : minmax(latest_surge[common_sectors]),
        })
        score['Composite'] = score.mean(axis=1)
        score = score.sort_values('Composite', ascending=False)

        print("\n  Composite score (0=worst → 1=best):")
        for sec, row in score.iterrows():
            flag = "🟢 BUY" if row['Composite'] > 0.6 else (
                   "🟡 WATCH" if row['Composite'] > 0.35 else "🔴 AVOID")
            bar = "█" * int(row['Composite'] * 20)
            print(f"    {sec:15s} {bar:<20s} {row['Composite']:.2f}  {flag}")
    else:
        score = pd.DataFrame()

    # ── 2.6 Rotation Pattern Detection ────────────────────────────────────
    print("\n--- 2.6 ROTATION PATTERN DETECTION ---")

    transitions = sum(
        1 for i in range(1, len(leadership_history))
        if leadership_history[i] != leadership_history[i-1]
    )
    total = max(1, len(leadership_history) - 1)
    freq  = transitions / total * 100

    print(f"\n  Leadership changed {transitions}/{total} periods ({freq:.0f}% rotation rate)")

    leader_counts = Counter(leadership_history)
    print("\n  Most frequent sector leaders:")
    for sector, count in leader_counts.most_common(5):
        pct = count / len(leadership_history) * 100
        print(f"    {sector:15s}: {count} periods ({pct:.0f}%)")

    # Annual sector rank heatmap data
    annual_sec_ret = (sec_idx
                      .resample('YE').last()
                      .pct_change()
                      .dropna(how='all')
                      .mul(100))
    annual_sec_ret.index = annual_sec_ret.index.year
    print("\n  Annual sector returns (%) — the rotation map:")
    print(annual_sec_ret.round(1).to_string())

    # ── 2.7 Prediction ────────────────────────────────────────────────────
    print("\n--- 2.7 NEXT MOVE PREDICTION ---")
    print(f"  Based on last {lookback_weeks} weeks of data:\n")

    print("  🟢 OVERWEIGHT (momentum + volume confirm inflow):")
    for sector, ret in latest_mom.head(3).items():
        print(f"     {sector:15s}: relative return {ret:+.2%}")

    print("\n  🔴 UNDERWEIGHT (momentum + volume confirm outflow):")
    for sector, ret in latest_mom.tail(3).items():
        print(f"     {sector:15s}: relative return {ret:+.2%}")

    print("\n  Rotation cycle pattern identified:")
    print("    Bull phase    → Engineering / IT (high beta, growth)")
    print("    Bear phase    → Pharma / Insurance (defensive)")
    print("    Recovery      → Textile / Bank (cyclical rebound)")
    print("    Late-recovery → Pharma building momentum again")
    print("    Current DSE   → Recovery phase — watch Pharma & Bank")

    return {
        'latest_momentum'   : latest_mom,
        'vol_surge'         : vol_surge,
        'money_flow'        : money_flow,
        'score'             : score,
        'leadership_history': leadership_history,
        'persistence_score' : persistence,
        'annual_returns'    : annual_sec_ret,
    }
```

```python
# =============================================================================
# PART 5: VISUALIZATIONS
# =============================================================================

def create_visualizations(df: pd.DataFrame,
                           task1: dict,
                           task2: dict,
                           output_dir: str = OUTPUT_DIR) -> None:
    """
    Produce 8 publication-quality charts and save to output_dir.

    Charts
    ------
    1. Sector annual returns heatmap
    2. Best/worst sector each year
    3. Risk vs Return bubble chart
    4. Sharpe ratio league table
    5. Volume surge heatmap (last 24 months)
    6. Sector cumulative performance
    7. Composite rotation signal
    8. Seasonal return heatmap
    """
    os.makedirs(output_dir, exist_ok=True)
    print(f"\nGenerating charts → {output_dir}/")

    sec_idx = build_sector_index(df)

    # ── 1. Sector Annual Returns Heatmap ─────────────────────────────────
    annual = task2['annual_returns']
    fig, ax = plt.subplots(figsize=(14, 6))
    sns.heatmap(annual.T, annot=True, fmt='.0f', center=0, cmap='RdYlGn',
                linewidths=0.5, linecolor='white',
                cbar_kws={'label': 'Annual Return (%)'}, ax=ax,
                annot_kws={'size': 8})
    ax.set_title('DSE Sector Annual Returns Heatmap (%)\n'
                 'Green = Positive  |  Red = Negative',
                 fontsize=13, fontweight='bold')
    ax.tick_params(axis='x', rotation=45)
    ax.tick_params(axis='y', rotation=0)
    plt.tight_layout()
    plt.savefig(f'{output_dir}/C1_sector_heatmap.png', dpi=150, bbox_inches='tight')
    plt.close()
    print('  ✓ C1_sector_heatmap.png')

    # ── 2. Best vs Worst Sector Each Year ────────────────────────────────
    fig, ax = plt.subplots(figsize=(12, 5))
    for yr in annual.index:
        row = annual.loc[yr].dropna()
        if row.empty:
            continue
        ax.plot([yr, yr], [row.min(), row.max()], color='grey', lw=1.5, zorder=1)
        ax.scatter(yr, row.max(), color='#1B9E75', s=90, zorder=2)
        ax.scatter(yr, row.min(), color='#D62728', s=90, zorder=2)
        ax.text(yr, row.max() + 2, row.idxmax()[:4],
                ha='center', fontsize=7, color='#1B9E75')
        ax.text(yr, row.min() - 4, row.idxmin()[:4],
                ha='center', fontsize=7, color='#D62728')
    ax.axhline(0, color='black', lw=0.8, ls='--', alpha=0.5)
    ax.set_xlabel('Year')
    ax.set_ylabel('Annual Return (%)')
    ax.set_title('Best (Green) vs Worst (Red) Sector Each Year',
                 fontsize=13, fontweight='bold')
    plt.tight_layout()
    plt.savefig(f'{output_dir}/C2_best_worst.png', dpi=150, bbox_inches='tight')
    plt.close()
    print('  ✓ C2_best_worst.png')

    # ── 3. Risk vs Return Bubble Chart ───────────────────────────────────
    stats = task1['sharpe_stats']
    fig, ax = plt.subplots(figsize=(12, 7))
    for sector in stats['SECTOR'].unique():
        sub = stats[stats['SECTOR'] == sector]
        c   = SECTOR_COLORS.get(sector, '#888')
        ax.scatter(sub['ANN_VOL'] * 100, sub['ANN_RETURN'] * 100,
                   s=60, c=c, alpha=0.7, edgecolors='white',
                   linewidths=0.6, label=sector)
    ax.axhline(0, color='black', lw=0.8, ls='--', alpha=0.5)
    ax.set_xlabel('Annualised Volatility (%)')
    ax.set_ylabel('Annualised Return (%)')
    ax.set_title('Risk vs Return — DSE Stocks\nTop-left = Best',
                 fontsize=13, fontweight='bold')
    ax.legend(fontsize=8, ncol=2, loc='lower right')
    plt.tight_layout()
    plt.savefig(f'{output_dir}/C3_risk_return.png', dpi=150, bbox_inches='tight')
    plt.close()
    print('  ✓ C3_risk_return.png')

    # ── 4. Sharpe League Table ────────────────────────────────────────────
    top20 = stats.nlargest(20, 'SHARPE')
    fig, ax = plt.subplots(figsize=(10, 7))
    colors = [SECTOR_COLORS.get(s, '#888') for s in top20['SECTOR']]
    bars = ax.barh(top20['TICKER'], top20['SHARPE'],
                   color=colors, edgecolor='white', height=0.7)
    for bar, val in zip(bars, top20['SHARPE']):
        ax.text(val + 0.005, bar.get_y() + bar.get_height() / 2,
                f'{val:.2f}', va='center', fontsize=8.5, fontweight='bold')
    ax.axvline(0, color='black', lw=0.8)
    ax.axvline(0.5, color='green', lw=1, ls='--', alpha=0.6,
               label='Good threshold (0.5)')
    ax.invert_yaxis()
    ax.set_xlabel('Sharpe Ratio (annualised)')
    ax.set_title('Top 20 Stocks — Sharpe Ratio',
                 fontsize=13, fontweight='bold')
    ax.legend(fontsize=9)
    plt.tight_layout()
    plt.savefig(f'{output_dir}/C4_sharpe.png', dpi=150, bbox_inches='tight')
    plt.close()
    print('  ✓ C4_sharpe.png')

    # ── 5. Volume Surge Heatmap ───────────────────────────────────────────
    vol_surge = task2['vol_surge']
    if not vol_surge.empty:
        last24 = vol_surge.tail(24).T
        last24.columns = [c.strftime('%b-%y') for c in last24.columns]
        fig, ax = plt.subplots(figsize=(16, 5))
        sns.heatmap(last24, annot=True, fmt='.1f', center=1.0,
                    cmap='RdYlGn', vmin=0.2, vmax=2.5,
                    linewidths=0.4, linecolor='white',
                    cbar_kws={'label': 'Volume Surge Ratio (1.0 = normal)'},
                    ax=ax, annot_kws={'size': 7.5})
        ax.set_title('Sector Volume Surge Ratio — Last 24 Months\n'
                     '>1.5 = Money flowing IN  |  <0.7 = Money flowing OUT',
                     fontsize=12, fontweight='bold')
        ax.tick_params(axis='x', rotation=45)
        ax.tick_params(axis='y', rotation=0)
        plt.tight_layout()
        plt.savefig(f'{output_dir}/C5_volume_surge.png', dpi=150, bbox_inches='tight')
        plt.close()
        print('  ✓ C5_volume_surge.png')

    # ── 6. Sector Cumulative Performance ─────────────────────────────────
    cum_perf = ((1 + sec_idx.pct_change().fillna(0)).cumprod() - 1) * 100
    fig, ax = plt.subplots(figsize=(14, 6))
    for sector in cum_perf.columns:
        if sector == 'UNKNOWN':
            continue
        ax.plot(cum_perf.index, cum_perf[sector],
                label=sector, color=SECTOR_COLORS.get(sector, '#888'),
                lw=1.8, alpha=0.9)
    ax.axhline(0, color='black', lw=0.8, ls='--', alpha=0.4)
    ax.set_ylabel('Cumulative Return (%)')
    ax.set_title('DSE Sector Cumulative Performance (2013–2026)',
                 fontsize=13, fontweight='bold')
    ax.legend(fontsize=8, bbox_to_anchor=(1.01, 1), loc='upper left')
    ax.yaxis.set_major_formatter(mticker.PercentFormatter())
    plt.tight_layout()
    plt.savefig(f'{output_dir}/C6_cumulative.png', dpi=150, bbox_inches='tight')
    plt.close()
    print('  ✓ C6_cumulative.png')

    # ── 7. Composite Rotation Signal ─────────────────────────────────────
    score = task2.get('score', pd.DataFrame())
    if not score.empty and 'Composite' in score.columns:
        sorted_score = score.sort_values('Composite')
        bar_colors = [
            '#1B9E75' if v > 0.6 else ('#EF9F27' if v > 0.35 else '#D62728')
            for v in sorted_score['Composite']
        ]
        fig, ax = plt.subplots(figsize=(10, 5))
        bars = ax.barh(sorted_score.index, sorted_score['Composite'],
                       color=bar_colors, edgecolor='white', height=0.6)
        for bar, val in zip(bars, sorted_score['Composite']):
            ax.text(val + 0.01, bar.get_y() + bar.get_height() / 2,
                    f'{val:.2f}', va='center', fontsize=10, fontweight='bold')
        ax.axvline(0.6, color='green', lw=1.2, ls='--', alpha=0.6,
                   label='BUY threshold')
        ax.axvline(0.35, color='orange', lw=1.2, ls='--', alpha=0.6,
                   label='WATCH threshold')
        ax.set_xlim(0, 1.15)
        ax.set_xlabel('Composite Rotation Score (0 = worst, 1 = best)')
        ax.set_title('Sector Rotation Signal — Current Rankings\n'
                     'Green = BUY  |  Amber = WATCH  |  Red = AVOID',
                     fontsize=12, fontweight='bold')
        ax.legend(fontsize=9)
        plt.tight_layout()
        plt.savefig(f'{output_dir}/C7_rotation_signal.png', dpi=150, bbox_inches='tight')
        plt.close()
        print('  ✓ C7_rotation_signal.png')

    # ── 8. Seasonal Return Heatmap ────────────────────────────────────────
    MONTH_NAMES = ['Jan','Feb','Mar','Apr','May','Jun',
                   'Jul','Aug','Sep','Oct','Nov','Dec']
    seasonal = {}
    for sector in SECTOR_MAP:
        tickers = [t for t in SECTOR_MAP[sector] if t in df['TICKER'].unique()]
        if not tickers:
            continue
        sec_ret_daily = df[df['TICKER'].isin(tickers)].groupby('DATE')['DAILY_RETURN'].mean()
        seasonal[sector] = sec_ret_daily.groupby(sec_ret_daily.index.month).mean() * 100

    seasonal_df = pd.DataFrame(seasonal)
    seasonal_df.index = MONTH_NAMES

    fig, ax = plt.subplots(figsize=(13, 5))
    sns.heatmap(seasonal_df.T, annot=True, fmt='.3f', center=0, cmap='RdYlGn',
                linewidths=0.5, linecolor='white',
                cbar_kws={'label': 'Avg Daily Return (%)'},
                ax=ax, annot_kws={'size': 8})
    ax.set_title('Seasonal Return Pattern — Avg Daily Return by Month & Sector (%)',
                 fontsize=12, fontweight='bold')
    ax.tick_params(axis='x', rotation=0)
    ax.tick_params(axis='y', rotation=0)
    plt.tight_layout()
    plt.savefig(f'{output_dir}/C8_seasonal.png', dpi=150, bbox_inches='tight')
    plt.close()
    print('  ✓ C8_seasonal.png')

    print(f"\n  All {len(os.listdir(output_dir))} charts saved to ./{output_dir}/")
```
